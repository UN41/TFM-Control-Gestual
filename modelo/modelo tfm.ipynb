```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df1=pd.read_csv("/content/drive/MyDrive/tfm/datos_pose_mano1.csv", sep=";", decimal=",")
df2=pd.read_csv("/content/drive/MyDrive/tfm/datos_pose_mano2.csv", sep=";", decimal=",")
df3=pd.read_csv("/content/drive/MyDrive/tfm/datos_pose_mano3.csv", sep=";", decimal=",")
df4=pd.read_csv("/content/drive/MyDrive/tfm/datos_pose_mano4.csv", sep=";", decimal=",")
df5=pd.read_csv("/content/drive/MyDrive/tfm/datos_pose_mano5.csv", sep=";", decimal=",")
```

```python
df1_subset = df1.head(2000)
df2_subset = df2.head(2000)
df3_subset = df3.head(2000)
df4_subset = df4.head(2000)
df5_subset = df5.head(2000)

# Crea una lista con los DataFrames subset
```

```python
merged_df=pd.concat([df1_subset,df2_subset,df3_subset,df4_subset,df5_subset], axis=0)
merged_df
```

```python
X = merged_df.drop('pose', axis=1)  # Características (todas las columnas excepto 'pose')
y = merged_df['pose']  # Variable objetivo ('pose')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state =42)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],  # Número de árboles en el bosque
    'max_depth': [None, 5, 10],  # Profundidad máxima de cada árbol
    'min_samples_split': [2, 5, 10],  # Número mínimo de muestras requeridas para dividir un nodo
}
```

```python
# Crear el clasificador Random Forest
rf = RandomForestClassifier(random_state=42)

# Crear el objeto Grid Search
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)

# Entrenar el modelo con Grid Search
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Obtener el modelo con los mejores hiperparámetros
best_rf_model = grid_search.best_estimator_

# Evaluar el modelo en el conjunto de prueba
accuracy = best_rf_model.score(X_test, y_test)

# Imprimir los resultados
print("Mejores hiperparámetros:", best_params)
print("Mejor score durante la validación cruzada:", best_score)
print("Precisión en el conjunto de prueba:", accuracy)
```

```python
param_grid = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
```

```python
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Obtener el modelo con los mejores hiperparámetros
best_knn_model = grid_search.best_estimator_

# Evaluar el modelo en el conjunto de prueba
accuracy = best_knn_model.score(X_test, y_test)

# Imprimir los resultados
print("Mejores hiperparámetros:", best_params)
print("Mejor score durante la validación cruzada:", best_score)
print("Precisión en el conjunto de prueba:", accuracy)
```
